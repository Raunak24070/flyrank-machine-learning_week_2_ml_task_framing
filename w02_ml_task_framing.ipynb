## Why ML beats a fixed rule

A fixed rule such as "always rank the most popular product first" will not work for every search.

Different users type different queries, and different features affect relevance at the same time. Some products may be popular but not relevant to a specific query.

Machine learning can learn these patterns from historical data and combine many signals together, which is much more flexible than writing hundreds of if-else rules.
